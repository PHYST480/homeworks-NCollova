```python
'''
Repeat the simple MLE example from class (Inference2.ipynb), but with 2 changes.  
First allow for a variable number of data points (and show for `N=3` and `N=10`. 
Second, limit the range of valid solutions for $\mu$ to $\mu>1.1$ (just for `N=10`). 
How does that affect your results?  Use a random seed of 42 for all.
'''
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm
%matplotlib inline

#Copying over what we did in Interence2.ipynb
#for N=3:
np.random.seed(42) #initialize 'random state'
N = 3
mu = 1
sigma = 0.2 
sample = norm(mu,sigma).rvs(N)#randomly select N datapoints from normal distribution with selected mean, mu; and std, sigma
print(sample)

#plot
muGrid = np.linspace(0,2,1000) #range of potentially best mu's (or likelihoods!)

#could just loop through - but just copying and pasting what we did in class :)
L1 = norm.pdf(muGrid,loc=sample[0],scale=sigma)
L2 = norm.pdf(muGrid,loc=sample[1],scale=sigma)
L3 = norm.pdf(muGrid,loc=sample[2],scale=sigma)
L = L1*L2*L3

#plot now!
fig, ax = plt.subplots(figsize=(7, 5))
plt.plot(muGrid, L1, ls='-', c='green', label=r'$L(x_1)$')
plt.plot(muGrid, L2, ls='-', c='red', label=r'$L(x_2)$')
plt.plot(muGrid, L3, ls='-', c='blue', label=r'$L(x_3)$')
plt.plot(muGrid, L/N, ls='-', c='black', label=r'$L(\{x\})$')

plt.xlim(0.2, 1.8)
plt.ylim(0, 2.5)
plt.xlabel('$\mu$') #Leave out or adjust if no latex
plt.ylabel(r'$p(x_i|\mu,\sigma)$') #Leave out or adjust if no latex
plt.title('MLE for Gaussian Distribution')
plt.legend()

#print out maximum likelihood:
indices = np.argsort(L)
index_max = indices[-1]
print("Likelihood for N=3 is maximized at %.3f" % muGrid[index_max])
```

```python
'''
Produce the "trace" plot for the clear/cloudy day problem from Inference2.ipynb.
Then plot the histogram of that trace, throwing away steps in the trace that could 
be considered part of the burn-in. Your plots should look like the ones from class (or nicer!) 
Confirm that you get the same answer for day  N  as day  N+1 
and it doesn't matter whether is was clear or cloudy on the day that you started.
'''
#From Inference2.ipynb produce the trace plot; Markov Chain for clear/cloudy day --> 
#Next state only depends on current state so define current states (cloudy or clear and today)
#1) Starting at a random position, evaluate the likelihood
%matplotlib inline

import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

#Define probabilities from notebook for probability that tomorrow will be clear from today:
#Starting at a random position, evaluate the likelihood

#assign 0 to be cloudy and 1 to be clear
clear = 1 #initialize clear because looping through to obtain probability of clear days tomorrow
today = 0 #start at 0 from today before looping through what tomorrow will be 

#create empty array of markov chain that can be appended to after looping through probabilities
markov = np.array([])
#create empty array of probclear which is the probability of it being clear tomorrow because going to loop through N and create array!
probclear = np.array([])

N = 10000
days = np.arange(N)

#for i some range of number values for 'days' axis
for i in days:
    p = np.random.random(1)#start random walk through the probabilities with a random position
    
    #likelihoods of today's chances - either cloudy or clear
    if (today): #this means if it's clear today, probability of it being cloudy or clear tomorrow 
        if (p<=0.1):
            tomorrow = 0
        else:
            tomorrow = 1 
            
    else: #this means if it's cloudy today, probability of it being clear or cloudy tomorrow
        if (p<=0.5):
            tomorrow = 0 #0 chance that tomorrow will also be cloudy, etc.
        else:
            tomorrow = 1 
    
    markov = np.append(markov,tomorrow) #append these values of tomorrow from random walk, markov empty array

    #now tomorrow because chain fig is plotting P(clear) vs. days so want to count probability of clear tomorrow's
    if (tomorrow): 
        clear = clear + 1
        
    #probclear.append(clear/(i+1))
    probclear = np.append(probclear,clear/(i+1))
    
    today = tomorrow #because tomorrow was calculated based on probabilities of today



plt.figure()
plt.title('Markov Chain Monte Carlo Trace for the Probability that Tomorrow will be a Clear Day')
plt.plot(days, probclear)
plt.xlabel('Days')
plt.ylabel('P(clear)')
plt.ylim(0.65,1)
```
